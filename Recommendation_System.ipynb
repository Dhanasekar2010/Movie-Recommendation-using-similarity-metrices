```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import sigmoid_kernel
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df=pd.read_csv("netfilx_datasets.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
tfidf=TfidfVectorizer(stop_words="english")
tfidf1=tfidf.fit_transform(df.description)
```

```python
similarities=sigmoid_kernel(tfidf1,tfidf1)
pd.DataFrame(similarities)
```

```python
similarities.shape
```

```python
indices=pd.Series(df.index,index=df.title)
indices
```

```python
def recommended_movie(similar_movie):
  if similar_movie in indices.index:
    index=np.where(similar_movie==indices.index)[0][0]
    similar=sorted(list(enumerate(similarities[index])),reverse=True,key=lambda x:x[1])[1:6]
    for movie in similar:
      print(indices.index[movie[0]])
  else:
    print("Movie is not in list")
```

```python
recommended_movie("Blood & Water")
```
